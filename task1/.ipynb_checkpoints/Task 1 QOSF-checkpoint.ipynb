## Part 1

All one qubit states are part of a complex hilbert space of two dimensions; because $|0\rangle, |1\rangle$ are the basis vectors. All transformations in this hilbert space can be carried out by unitary matrices from $SU(2)$ and its equivalent $SO(3)$ vector spaces. The elements of $SO(3)$ are the rotations $Rx$,$Ry$ and $Rz$, along the x, y and z axis. So a most general one qubit state can be produced by these three rotations. This is not surprising because the $Rx$ rotation covers points in the $y-z$ plane, the $Ry$ rotation covers points in the $x-z$ plane and the $Rz$ rotation covers points in the $x-y$ plane, together they encompass the entire bloch sphere.

The function `oneQubitParameterizedCircuit` defined below defines this general one qubit parameterized circuit and then we show that this circuit can reach any point on the bloch sphere. Qiskit does not have the ability to plot multiple points on the bloch sphere, so we will use `qutip`, we will convert the statevectors as numpy arrays to be plotted with qutip.

```python
def plot_bloch_single_qubit_states(bloch_data, title="Generic Single Qubit states", figsize=(10,10), point_size=25):
    B = Bloch(figsize=figsize)
    B.point_size = [point_size]
    vectors = np.array([_bloch_multivector_data(datum)[0] for datum in bloch_data])
    B.add_points([vectors[:,0], vectors[:,1], vectors[:,2]])
    
    B.show()
```

```python
def oneQubitParameterizedCircuit(name: str,
                                 theta_x: qiskit.circuit.Parameter,
                                 theta_y: qiskit.circuit.Parameter, 
                                 theta_z: qiskit.circuit.Parameter) -> qiskit.QuantumCircuit:
    circuit = QuantumCircuit(1, name=name)
    circuit.rx(theta_x, 0)
    circuit.ry(theta_y, 0)
    circuit.rz(theta_z, 0)
    return circuit
```

```python
theta_x, theta_y, theta_z = Parameter('theta_x'), Parameter('theta_y'), Parameter('theta_z')
circuit = oneQubitParameterizedCircuit('test', theta_x, theta_y, theta_z)
```

Since we start with the state $|0\rangle$ we chose not to start the gate sequence with the $Rz$ gate as this will only change the overall phase.

```python
circuit.draw('mpl')
```

```python
def sample_one_qubit_states_on_bloch_sphere(samples=10):
    
    sampled_theta_x = np.linspace(0, 2 * np.pi, samples)
    sampled_theta_y = np.linspace(0, 2 * np.pi, samples)
    sampled_theta_z = np.linspace(0, 2 * np.pi, samples)
    
    theta_x, theta_y, theta_z = Parameter('theta_x'), Parameter('theta_y'), Parameter('theta_z')
    qc = oneQubitParameterizedCircuit('test', theta_x, theta_y, theta_z)
    
    circuits = [qc.bind_parameters({theta_x: val_x, theta_y: val_y, theta_z: val_z}) for val_x in sampled_theta_x
                for val_y in sampled_theta_y for val_z in sampled_theta_z]
    
    states = [Statevector.from_instruction(circuit) for circuit in circuits]
    plot_bloch_single_qubit_states(states)
```

## Part 2

### Swap Test:

As defined in the wiki for [swap test](https://en.wikipedia.org/wiki/Swap_test), if we have two states $\phi$ and $\psi$, we measure the ancilla to be 0 with probability $p_0 = \frac{1}{2}+\frac{|\langle\psi|\phi\rangle|^2}{2}$. So now $|\langle\psi|\phi\rangle|^2 = 2p_0 - 1$, $2p_0 - 1$ is zero when $\phi$ and $\psi$ are orthogonal and $2p_0 - 1$ is 1, when these two are equal. So to maximize similarity we have to maximize $2p_0 - 1$ or minimize $1 - 2p_0$. Therefore $1-2p_0$ becomes our objective function or loss.

```python
class OneQubitSwapTest:
    def __init__(self, psi: qiskit.QuantumCircuit):
        assert(psi.num_qubits is 1)
        self.psi = psi
        self.simulator = Aer.get_backend('qasm_simulator')
    
    def cswap(self, phi: qiskit.QuantumCircuit) -> qiskit.QuantumCircuit:
        assert(phi.num_qubits is 1)
        qr = QuantumRegister(3)
        cr = ClassicalRegister(1)
        qc = QuantumCircuit(qr, cr)
        qc = qc.compose(self.psi, qubits=[0])
        qc = qc.compose(phi, qubits=[1])
        qc.barrier()
        qc.h(2)
        qc.cswap(qr[2], qr[0], qr[1])
        qc.h(2)
        qc.barrier()
        qc.measure(2, 0)
        return qc
    
    def oneQubitAnsatz(self, name: str, theta_x: qiskit.circuit.Parameter,
                       theta_y: qiskit.circuit.Parameter,
                       theta_z: qiskit.circuit.Parameter) -> qiskit.QuantumCircuit:
        circuit = QuantumCircuit(1, name=name)
        circuit.rx(theta_x, 0)
        circuit.ry(theta_y, 0)
        circuit.rz(theta_z, 0)
        return circuit
    
    def _loss(self, params):
        return self.loss(params[0], params[1], params[2])

    def loss(self, r_x, r_y, r_z, shots=1000):
        theta_x, theta_y, theta_z = Parameter('theta_x'), Parameter('theta_y'), Parameter('theta_z')
        qc = self.oneQubitAnsatz('test', theta_x, theta_y, theta_z)
        swap = self.cswap(qc)
        swap = swap.bind_parameters({theta_x: r_x, theta_y: r_y, theta_z: r_z})
        result = execute(swap, self.simulator, shots=shots).result()
        counts = result.get_counts(swap)
        prob_0 = counts['0']/shots
        return (1 - 2 * prob_0)

    def plotLossLandscape(self, samples=10):
        sampled_theta_x = np.linspace(0, 2 * np.pi, samples)
        sampled_theta_y = np.linspace(0, 2 * np.pi, samples)
        sampled_theta_z = np.linspace(0, 2 * np.pi, samples)
        
        fig, (ax1, ax2, ax3) = plt.subplots(1,3, constrained_layout=True, figsize=(15, 5))
        
        # plot x-y variation
        x_y_losses = np.array([self.loss(theta_x, theta_y, 0) for theta_x in sampled_theta_x for theta_y in sampled_theta_y])
        x_y_losses = x_y_losses.reshape(samples, samples)
        cs = ax1.contourf(sampled_theta_x, sampled_theta_y, x_y_losses, 20, cmap='RdGy')
        ax1.set_title("Rx and Ry")
        fig.colorbar(cs, ax=ax1, shrink=0.9)
        
        # plot x-z variation
        x_z_losses = np.array([self.loss(theta_x, 0, theta_z) for theta_x in sampled_theta_x for theta_z in sampled_theta_z])
        x_z_losses = x_z_losses.reshape(samples, samples)
        cs = ax2.contourf(sampled_theta_x, sampled_theta_z, x_z_losses, 20, cmap='RdGy')
        ax2.set_title("Rx and Rz")
        fig.colorbar(cs, ax=ax2, shrink=0.9)

        # plot y-z variation
        y_z_losses = np.array([self.loss(0, theta_y, theta_z) for theta_y in sampled_theta_y for theta_z in sampled_theta_z])
        y_z_losses = y_z_losses.reshape(samples, samples)
        cs = ax3.contourf(sampled_theta_y, sampled_theta_z, y_z_losses, 20, cmap='RdGy')
        ax3.set_title("Ry and Rz")
        fig.colorbar(cs, ax=ax3, shrink=0.9)
    
        plt.suptitle("Value of objective function after varying each of Rx, Ry and Rz between 0 and 2pi")
        plt.show()
        return
    
    def optimize(self, method='L-BFGS-B'):
        minimized_angles = minimize(self._loss, np.array([0, 0, 0]), method=method, bounds=((0, 2*np.pi), (0, 2*np.pi), (0, 2*np.pi)), options={'maxiter':1000})
        return minimized_angles['fun']
```

Now we are left with the complex task of choosing an optimizer. As mentioned in the paper *Classical Optimizers for Noisy Intermediate-Scale quantum Devices* by Lavrijsen et al. A good optimizer will be less susceptible to noise related erros and scaled well with the number of qubits. Since we are running only on optimizers noise is not a problem and since we are only concerned with a single qubit circuit with three parameters, we will choose an optimizer that converges well. Handpick a bunch of optimizers with no justification and check which one of them leads to least error.

```python
 def optimizersGraph():
        optimizers = ['Nelder-Mead', 'Powell', 'BFGS', 'L-BFGS-B', 'COBYLA']
        random_params = np.linspace(0, 2 * np.pi, 9)
        sampled_theta_x, sampled_theta_y, sampled_theta_z = random_params[:3], random_params[3:6], random_params[6:9]
        losses = [0] * len(optimizers)
        
        theta_x, theta_y, theta_z = Parameter('theta_x'), Parameter('theta_y'), Parameter('theta_z')
        qc = oneQubitParameterizedCircuit('test', theta_x, theta_y, theta_z)
        circuits = [qc.bind_parameters({theta_x: val_x, theta_y: val_y, theta_z: val_z}) for val_x in sampled_theta_x
                for val_y in sampled_theta_y for val_z in sampled_theta_z]
    
        for idx, optimizer in enumerate(optimizers):
            loss = 0
            for circuit in circuits:
                oneQubitSwapTest = OneQubitSwapTest(circuit)
                loss += abs(oneQubitSwapTest.optimize(method=optimizer))
            losses[idx] = loss/len(circuits)
        
        x = np.arange(len(optimizers))
        plt.bar(x, height=losses)
        plt.xticks(x, optimizers)
```

```python
optimizersGraph()
```

From the graph above we see that **L-BFGS-B** gives the best convergence.

## Part 3

Algorithm that performs grid search on a n qubit product state:

1. We assume that the state we are checking is a product state.
2. given n qubit state to check, we iterate over all the product states that are possible with n qubits.
3. Compose a new circuit with the n qubit state thats given and the product state that you are iterating over with an additional ancilla. 
4. Apply a h gate to the ancilla
5. For each of these states:
    * Iterate over each of the qubits of the two states and use cswap with the control being the ancilla.
6. Apply a h gate to the ancilla
7. Measure the value of the ancilla, if we get a a majority of measurements a 0 then both the states are equal and output this bit representation of this state.

```python
# Takes two states as input and returns if the two states are equivalent or not. 
# Returns True if they are equivalent else False.
def productStatesTest(psi, phi):
    n = psi.num_qubits
    ancilla = 2 * n
    qr = QuantumRegister(2*n + 1)
    cr = ClassicalRegister(1)
    qc = QuantumCircuit(qr, cr)
    qc = qc.compose(psi, qubits=list(range(n)))
    qc = qc.compose(phi, qubits=list(range(n, 2*n)))
    
    qc.h(ancilla)
    for i in range(n):
        qc.barrier()
        j = n + i
        qc.cswap(qr[ancilla], qr[i], qr[j])
        qc.barrier()
    qc.h(ancilla)
    qc.measure(ancilla, 0)
    
    shots = 1000
    result = execute(qc, Aer.get_backend('qasm_simulator'), shots=shots).result()
    counts = result.get_counts(qc)
    prob_0 = counts['0']/shots
    prob_1 = counts['1']/shots
    return True if prob_0 > prob_1 else False
```

```python
qr = QuantumRegister(3)
psi = QuantumCircuit(qr)
qr2 = QuantumRegister(3)
phi = QuantumCircuit(qr2)
phi.x(0)
productStatesTest(psi, phi)
```

```python
def checkProductStates(psi):
    n = psi.num_qubits    
    binary_combinations = list(itertools.product([0, 1], repeat=n))
    
    qr = QuantumRegister(n)
    circuits = []
    
    for combination in binary_combinations:
        qc = QuantumCircuit(qr)
        for idx, qubit in enumerate(combination):
            if qubit == 1:
                qc.x(idx)
        if productStatesTest(psi, qc):
            return ''.join(str(e) for e in combination)
```
